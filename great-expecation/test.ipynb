```python
import great_expectations as ge
import pandas as pd
```

```python
context = ge.data_context.DataContext('/Users/rajsingh/work/repos/github/python-learning/great-expecation/gx/')
```

```python
data = {
    'column1': [1,2,3,4,5],
    'column2': ['a','b','c','d','e']
}

df = pd.DataFrame(data)
df
```

```python
expectation_suite_name = "my_expectation_suite"
context.create_expectation_suite(expectation_suite_name)
```

```python
# Create Expecatation for the DataFrame 
batch = ge.dataset.PandasDataset(df)
batch.expect_column_values_to_be_in_set('column1', [1,2,3,4,5])
batch.expect_column_values_to_be_match_regex('column2', r'[a-e]')
```

```python
context.save_expectation_suite(batch.get_expectation_suite(), expectation_suite_name)
```

```python
result = context.run_validation_operator(
    'action_list_operator',
    run_id='batch_'+ge.data_asset.util.identifier.batch_identifier(batch)
)
```
