```python

# Slicing 
movie='The Godfather'
print(movie[0:2])

# string_variable[start:stop]
print(movie[2:5])
print(movie[:2])
print(movie[4:])
print(movie[-2:])

# movie[:i] + movie[1:] is euqal to movie
print(movie[:4] + movie[4:])
print(movie[:43])
print(movie[3:100])

# string_variable[start:stop: step]
print(movie[0:10:2])
print(movie[::])

# Reverse Order
print(movie[::-1])

# s[::-1][::-1] == s


digits = '0123456789'
#using string slicing, create a variable called x that stores the substring '9630'

result = digits[0::3][::-1]
print(result)


result = digits[::-3]
print(result)
```
